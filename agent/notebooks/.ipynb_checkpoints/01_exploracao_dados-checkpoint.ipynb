```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from pathlib import Path
import warnings

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", 60)
pd.set_option("display.float_format", "{:.2f}".format)

# Paleta de cores do projeto
CORES = {
    "brasileirao_a": "#009C3B",
    "premier_league": "#3D195B",
    "la_liga":        "#EE1B2D",
}
sns.set_theme(style="darkgrid", palette="muted", font_scale=1.05)
plt.rcParams.update({"figure.dpi": 110, "axes.titlesize": 13, "axes.labelsize": 11})

print("✅ Bibliotecas carregadas")
```

## ⚙️ Configuração — troque aqui para usar o banco real
Altere `BASE_DIR` para apontar para a raiz do projeto no seu computador.


```python
# ──────────────────────────────────────────────────────────────────────────────
# SWAP: quando o banco real estiver disponível, apenas atualize BASE_DIR
# para o caminho real do projeto, ex: Path("/Users/cristiano/oscabet-agent")
# ──────────────────────────────────────────────────────────────────────────────
BASE_DIR  = Path("../..").resolve()           # raiz do projeto (oscabet/)
RAW_DIR   = BASE_DIR / "data" / "raw"
PROC_DIR  = BASE_DIR / "data" / "processed"

MATCHES_PATH = RAW_DIR / "matches.csv"
STATS_PATH   = RAW_DIR / "match_stats.csv"

assert MATCHES_PATH.exists(), f"❌ Arquivo não encontrado: {MATCHES_PATH}\nExecute generate_data.py primeiro."
assert STATS_PATH.exists(),   f"❌ Arquivo não encontrado: {STATS_PATH}"

matches = pd.read_csv(MATCHES_PATH, parse_dates=["date"])
stats   = pd.read_csv(STATS_PATH)

print(f"matches.csv    → {len(matches):,} linhas × {matches.shape[1]} colunas")
print(f"match_stats.csv → {len(stats):,} linhas × {stats.shape[1]} colunas")
```

## 1. Visão Geral do Dataset

```python
print("── matches.csv ──────────────────────────────────────")
print(matches.dtypes.to_string())
print()
print(matches.head(3).to_string())
```

```python
print("── match_stats.csv (primeiras 10 colunas) ──────────")
print(stats.iloc[:, :10].dtypes.to_string())
print()
print(stats.iloc[:3, :10].to_string())
```

```python
# Sumário numérico das estatísticas principais
cols_show = [
    "home_shots", "away_shots", "home_shots_on_target", "away_shots_on_target",
    "home_corners", "away_corners", "total_corners",
    "home_yellow_cards", "away_yellow_cards", "total_yellow_cards",
    "home_xg", "away_xg", "home_possession",
]
stats[cols_show].describe().T.round(2)
```

## 2. Cobertura Temporal

```python
matches["year"] = matches["date"].dt.year

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Jogos por ano
year_counts = matches.groupby("year").size()
axes[0].bar(year_counts.index.astype(str), year_counts.values,
            color=sns.color_palette("Blues_d", len(year_counts)))
axes[0].set_title("Partidas por Ano"); axes[0].set_ylabel("Qtde")
for rect, v in zip(axes[0].patches, year_counts.values):
    axes[0].text(rect.get_x()+rect.get_width()/2, rect.get_height()+10,
                 str(v), ha="center", va="bottom", fontsize=9)

# Jogos por temporada × liga
pivot = matches.groupby(["season","league"]).size().unstack(fill_value=0)
pivot.plot(kind="bar", ax=axes[1], color=[CORES[c] for c in pivot.columns], width=0.75)
axes[1].set_title("Partidas por Temporada e Liga")
axes[1].set_xlabel(""); axes[1].legend(title="Liga")
axes[1].tick_params(axis="x", rotation=30)

plt.tight_layout()
plt.show()
```

## 3. Distribuição de Resultados

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Geral
res_counts = matches["result"].value_counts().reindex(["H","D","A"])
colors_res = ["#2196F3","#9E9E9E","#F44336"]
axes[0].bar(["Mandante (H)","Empate (D)","Visitante (A)"],
            res_counts.values, color=colors_res, width=0.55)
for rect, v in zip(axes[0].patches, res_counts.values):
    pct = 100 * v / len(matches)
    axes[0].text(rect.get_x()+rect.get_width()/2, rect.get_height()+30,
                 f"{pct:.1f}%", ha="center", fontsize=10)
axes[0].set_title("Distribuição de Resultados (Geral)")
axes[0].set_ylabel("Partidas")

# Por liga
res_by_league = (matches.groupby(["league","result"]).size()
                 .unstack(fill_value=0)[["H","D","A"]]
                 .div(matches.groupby("league").size(), axis=0) * 100)
res_by_league.plot(kind="bar", ax=axes[1], color=colors_res, width=0.65, rot=0)
axes[1].set_title("Resultados por Liga (%)")
axes[1].set_ylabel("%"); axes[1].legend(["Mandante","Empate","Visitante"])
axes[1].yaxis.set_major_formatter(mticker.PercentFormatter())

plt.tight_layout(); plt.show()

print("\nResumo:")
print(res_counts.rename({
    "H":"Mandante vence","D":"Empate","A":"Visitante vence"}).to_string())
```

## 4. Distribuição de Gols

```python
df_full = matches.merge(stats[["match_id","home_xg","away_xg"]], on="match_id", how="left")

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

for ax, col, label, color in zip(
    axes,
    ["home_score", "away_score"],
    ["Gols Mandante", "Gols Visitante"],
    ["#2196F3", "#F44336"]
):
    counts = matches[col].value_counts().sort_index()
    ax.bar(counts.index, counts.values, color=color, alpha=0.8)
    ax.set_title(label); ax.set_xlabel("Gols"); ax.set_ylabel("Freq.")
    mean_val = matches[col].mean()
    ax.axvline(mean_val, color="black", ls="--", lw=1.5, label=f"Média={mean_val:.2f}")
    ax.legend()

# xG vs Gols reais
axes[2].scatter(df_full["home_xg"], df_full["home_score"],
                alpha=0.1, s=8, color="#2196F3", label="Mandante")
axes[2].scatter(df_full["away_xg"], df_full["away_score"],
                alpha=0.1, s=8, color="#F44336", label="Visitante")
lim = max(df_full[["home_xg","away_xg"]].max())
axes[2].plot([0, lim], [0, lim], "k--", lw=1, label="xG = Gols")
axes[2].set_title("xG vs Gols Reais"); axes[2].set_xlabel("xG")
axes[2].set_ylabel("Gols"); axes[2].legend()

plt.tight_layout(); plt.show()
```

## 5. Cartões Amarelos por Liga

```python
df_y = matches.merge(stats[["match_id","total_yellow_cards",
                             "home_yellow_cards","away_yellow_cards"]],
                     on="match_id", how="inner")

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Histograma por liga
for league, color in CORES.items():
    sub = df_y[df_y["league"] == league]["total_yellow_cards"]
    axes[0].hist(sub, bins=range(0, 16), alpha=0.6, color=color,
                 label=league.replace("_"," ").title(), density=True)
axes[0].set_title("Distribuição de Cartões Amarelos por Liga")
axes[0].set_xlabel("Total de Amarelos na Partida"); axes[0].set_ylabel("Densidade")
axes[0].legend()

# Médias por liga
means = df_y.groupby("league")[["home_yellow_cards","away_yellow_cards"]].mean()
means.rename(columns={"home_yellow_cards":"Mandante","away_yellow_cards":"Visitante"},
             inplace=True)
means.plot(kind="bar", ax=axes[1], color=["#2196F3","#F44336"], width=0.6, rot=0)
axes[1].set_title("Média de Amarelos por Liga (Mandante vs Visitante)")
axes[1].set_ylabel("Média"); axes[1].set_xlabel("")
axes[1].legend()

plt.tight_layout(); plt.show()

print("\nMédia de amarelos totais por liga:")
print(df_y.groupby("league")["total_yellow_cards"].agg(["mean","median","std"]).round(2).to_string())
```

## 6. Escanteios por Liga

```python
df_c = matches.merge(stats[["match_id","total_corners",
                             "home_corners","away_corners"]],
                     on="match_id", how="inner")

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

for league, color in CORES.items():
    sub = df_c[df_c["league"] == league]["total_corners"]
    axes[0].hist(sub, bins=range(0, 25), alpha=0.6, color=color,
                 label=league.replace("_"," ").title(), density=True)
axes[0].set_title("Distribuição de Escanteios por Liga")
axes[0].set_xlabel("Total de Escanteios"); axes[0].set_ylabel("Densidade")
axes[0].legend()

# Boxplot
data_box = [df_c[df_c["league"]==l]["total_corners"].values for l in CORES]
bp = axes[1].boxplot(data_box, patch_artist=True,
                     labels=[l.replace("_"," ").title() for l in CORES])
for patch, color in zip(bp["boxes"], CORES.values()):
    patch.set_facecolor(color); patch.set_alpha(0.7)
axes[1].set_title("Boxplot de Escanteios por Liga")
axes[1].set_ylabel("Total de Escanteios")

plt.tight_layout(); plt.show()

# Thresholds de classificação (do .env do projeto)
CORN_LOW_MAX  = 7
CORN_HIGH_MIN = 12

for league in CORES:
    sub = df_c[df_c["league"]==league]["total_corners"]
    low  = (sub <= CORN_LOW_MAX).mean() * 100
    high = (sub >= CORN_HIGH_MIN).mean() * 100
    mid  = 100 - low - high
    print(f"{league:20s}  low≤{CORN_LOW_MAX}: {low:.1f}%  |  mid: {mid:.1f}%  |  high≥{CORN_HIGH_MIN}: {high:.1f}%")
```

## 7. Correlações entre Features Principais

```python
corr_cols = [
    "home_shots_on_target","away_shots_on_target",
    "home_corners","away_corners","total_corners",
    "home_yellow_cards","away_yellow_cards","total_yellow_cards",
    "home_possession","home_xg","away_xg",
]
df_corr = (
    matches[["match_id","home_score","away_score"]]
    .merge(stats[["match_id"]+corr_cols], on="match_id", how="inner")
)

corr_mat = df_corr.drop("match_id", axis=1).corr()

fig, ax = plt.subplots(figsize=(13, 9))
mask = np.triu(np.ones_like(corr_mat, dtype=bool))
sns.heatmap(corr_mat, mask=mask, annot=True, fmt=".2f",
            cmap="RdYlGn", center=0, vmin=-1, vmax=1,
            linewidths=0.4, annot_kws={"size": 8}, ax=ax)
ax.set_title("Correlação entre Variáveis Principais", pad=15)
plt.tight_layout(); plt.show()
```

## 8. Dados Faltantes (stats_collected)

```python
miss_by_league = (
    matches.groupby("league")
    .apply(lambda x: pd.Series({
        "total":    len(x),
        "com_stats": x["stats_collected"].sum(),
        "sem_stats": (~x["stats_collected"].astype(bool)).sum(),
        "% cobertura": x["stats_collected"].mean() * 100,
    }))
    .round(1)
)
print(miss_by_league.to_string())

fig, ax = plt.subplots(figsize=(8, 4))
miss_by_league[["com_stats","sem_stats"]].plot(
    kind="bar", ax=ax, color=["#4CAF50","#F44336"], width=0.65, rot=0
)
ax.set_title("Cobertura de Estatísticas por Liga")
ax.set_ylabel("Partidas"); ax.legend(["Com stats","Sem stats"])
plt.tight_layout(); plt.show()
```
